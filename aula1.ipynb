```python
import pandas as pd

clientes = pd.read_csv('clientes.csv')
pedidos = pd.read_csv('pedidos.csv')

# Como identificar campos vazios dentro de uma planilha
clientes.isnull().sum()  # Retorna a contagem de valores nulos em cada coluna
# Como identificar campos vazios dentro de uma planilha
pedidos.isnull().sum()  # Retorna a contagem de valores nulos em cada coluna

# para preencher as celulas que possuem valores vazios com base em uma coluna
pedidos['custo_envio'] = pedidos['custo_envio'].fillna(999)
pedidos.isnull().sum()

# Preencher valores vazios com a média da coluna
pedidos['tempo_entrega_dias'] = pedidos['tempo_entrega_dias'].fillna(pedidos['tempo_entrega_dias'].mean())
pedidos.isnull().sum()

# caso queira remover as linhas que possuem valores vazios 
pedidos = pedidos.dropna(subset=['custo_envio'])
pedidos.isnull().sum()

# caso queira remover as linhas que possuem valores vazios em qualquer coluna
pedidos = pedidos.dropna()
pedidos.isnull().sum()

# encontrar valores duplicados em uma planilha
clientes.duplicated('email').sum() # Retorna a contagem de linhas duplicadas
clientes.describe() # Estatísticas descritivas das colunas numéricas

# remover valores duplicados em uma planilha
clientes = clientes.drop_duplicates('email')
clientes.describe() # Estatísticas descritivas das colunas numéricas

# salve as alterações em um noivo arquivo CSV
clientes.to_csv('clientes_limpo.csv', index=False) 
```

```python
import pandas as pd

# desafio

# Mostre quais colunas possuem valores nulos
students = pd.read_csv('./students.csv')   
students.isnull().sum()

# Remova todos os registros de email duplicados e depois mostre a quantidade
#  novamente quais colunas possuem valores nulos
students = students.drop_duplicates(subset=['email'])
students.isnull().sum()

# Preencha as colunas vazias de "idade" com a idade media. depois mostre quais colunas possuem valores vazios
students['idade'] = students['idade'].fillna(students['idade'].mean())
students.isnull().sum()
```
