```python
#Leituras de dados a partir de CSV, XLSX E TXT
import pandas as pd

# como carregar arquivos CSV
clientes = pd.read_csv('./clientes.csv') #dataframe
clientes.head(10) # ler as primeiras x linhas
clientes[["nome", "email", "nivel_fidelidade"]].head(10) # ler x quantidade de determinadas colunas
clientes.tail(10) # ler as últimas x linhas
clientes[["nome", "email", "nivel_fidelidade"]].head(10) # ler x quantidade de determinadas colunas

# como carregar arquivos XLSX
clientes_xlsx = pd.read_excel('./clientes.xlsx') #dataframe
clientes_xlsx.head(10) # ler as primeiras x linhas
clientes_xlsx[["nome", "email", "nivel_fidelidade"]].head(10) # ler x quantidade de determinadas colunas
clientes_xlsx.tail(10) # ler as últimas x linhas   
clientes_xlsx[["nome", "email", "nivel_fidelidade"]].head(10) # ler x quantidade de determinadas colunas

# como carregar arquivos TXT
clientes_txt = pd.read_csv('./clientes.txt', delimiter=';') #dataframe
clientes_txt.head(10) # ler as primeiras x linhas
clientes_txt[["nome", "email", "nivel_fidelidade"]].head(10) # ler x quantidade de determinadas colunas
clientes_txt.tail(10) # ler as últimas x linhas
clientes_txt[["nome", "email", "nivel_fidelidade"]].head(10) # ler x quantidade de determinadas colunas

```
